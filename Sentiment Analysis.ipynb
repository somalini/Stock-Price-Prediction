```python
import pandas as pd
```

```python
#Step 2- Read Data from compu|ter 
df = pd.read_csv("Combined_News_DJIA.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df = df.dropna()
df
```

```python
df.Label = df["Label"].astype(int)
```

```python
df = df.set_index(keys = df.Date)
```

```python
df
```

```python
df.drop(['Date'], axis = 1, inplace = True)
```

```python
#Step 3 - Remove the punctuation marks.
df.replace("[^a-zA-Z]"," ", regex = True, inplace = True)
df
```

```python
df.replace("b ","", regex = True, inplace = True)
df
```

```python
X = df.iloc[:,1:27]
X
```

```python
X
```

```python
X.columns
```

```python
#Step 4 - Convert all charactor to lower case.
for col in X.columns:
    X[col] = X[col].str.lower()
X
```

```python
#Step 5 - prepare train and test dataset
#print(X.shape)
X_train = X.iloc[:1500,:]
Y_train = df["Label"][:1500]

X_test = X.iloc[1500:,:]
Y_test = df["Label"][1500:]

print(X_train.shape)
print(X_test.shape)
```

```python
#Step 6 - Illerate each and every sentence in train dataset and add them together.
train_headlines = []
for i in range(len(X_train.index)):
    text = " ".join(str(x) for x in X_train.iloc[i, 0:25 ])
    train_headlines.append(text)
```

```python
train_headlines[0]
```

```python
#Step 7 - convert all the sentences into vectors by using count vectorizer
#count vectorizer helps you to build bag of words

from sklearn.feature_extraction.text import CountVectorizer
countvector = CountVectorizer(ngram_range = (2,2))
X_train_dataset = countvector.fit_transform(train_headlines)
X_train_dataset
```

```python
#Step 7: TF-IDF
'''from sklearn.feature_extraction.text import TfidfVectorizer
tfidfvectorizer = TfidfVectorizer(analyzer='word',stop_words = 'english')
X_train_dataset = tfidfvectorizer.fit_transform(train_headlines)
'''
```

```python
for i in X_train_dataset:
    print(i)   
```

```python
#Step 8 - implement Randomforest Classifier
from sklearn.ensemble import RandomForestClassifier
randomclassifier = RandomForestClassifier(n_estimators = 200, criterion='entropy')

randomclassifier.fit(X_train_dataset, Y_train)
```

```python
#step 9 - Itlerate each and every sentence in test dataset and add them together.
test_headlines = []
for i in range(len(X_test.index)):
    test_headlines.append(" ".join(str(x) for x in X_test.iloc[i, 0:25 ]))
len(test_headlines)
```

```python
test_dataset = countvector.transform(test_headlines)
test_dataset.shape
```

```python
'''tfidfvectorizer = TfidfVectorizer(analyzer='word',stop_words= 'english')
test_dataset = tfidfvectorizer.fit_transform(test_headlines)'''
```

```python
pred = randomclassifier.predict(test_dataset)
```

```python
X_test
```

```python
print(pred)
print(pred.shape)
```

```python
actural_data = [i for i in Y_test]
print(actural_data)
print(len(actural_data))
```

```python
#Step 12 - Apply Confusion Matrix 
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(actural_data, pred)
matrix
```

```python
#Step 13 - classification report 
from sklearn.metrics import classification_report
report = classification_report(actural_data,pred)
print(report)
```

```python
#Step 14 - Check Model Accuracy
from sklearn.metrics import accuracy_score
score = accuracy_score(Y_test,pred)
print(score * 100)
```
